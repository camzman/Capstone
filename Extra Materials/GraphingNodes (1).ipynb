```python
# libraries
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Create a Dictionary of all the Orders
# These are your coordinate points! And each coordinate is given a name (which is a number 1 through 22. 
# That number shows up in the black node of the plot)
pos = {1: (0,0), 
2: (86,26), 
3: (28,56), 
4: (91,40), 
5: (70,90), 
6: (84,57), 
7: (48,71), 
8: (61,65), 
9: (20,40), 
10: (32,59), 
11: (63,93), 
12: (46,57), 
13: (90,42), 
14: (71,59), 
15: (52,41), 
16: (3,64), 
17: (94,30), 
18: (52,42), 
19: (49,18), 
20: (16,84), 
21: (33,36), 
22: (0,0)
}
```

```python
# Add your Truck Route
# These are the arcs between coordinate points! 
G_1 = nx.Graph()
truckedgelist =  [[1,9], # this point draws a line from order point 1 (which is coordinate (0,0)) to order point 9 (coordinate (20,40))
[3,10], 
[4,18], 
[6,13], 
[7,8], 
[8,14], 
[9,3], 
[10,12], 
[12,7], 
[13,4], 
[14,6], 
[15,21], 
[18,15], 
[21,22]
]
```

```python
# Add your Drone Routes
# Just another set of arcs.  Defined these separately so that I could make them a different color
droneedgelist = [[2,13], 
[3,16], 
[4,17], 
[5,8], 
[7,11], 
[8,5], 
[10,20], 
[11,7], 
[13,2], 
[15,19], 
[16,3], 
[17,4], 
[19,15], 
[20,10]
]
```

```python
# Plot the Routes

for (u,v) in truckedgelist:
    G_1.add_edge(v, u, capacity=1) 

    
for (u,v) in droneedgelist:
    G_1.add_edge(v, u, capacity=1) 

    
fig, ax = plt.subplots(figsize=(10,10))
nx.draw(G_1, pos, edgelist=truckedgelist, edge_labels=True, with_labels=True, font_color='w', node_color='black', ax=ax)
nx.draw_networkx_edges(G_1, pos, edgelist=droneedgelist, edge_color='darkorange', style='dotted', ax=ax)
limits=plt.axis('on')
ax.tick_params(left=True, bottom=True, labelleft=True, labelbottom=True)
plt.show()
#plt.savefig("route-sim1-k1.png")
```
